<h3>Csv Input</h3>

```python
import pandas as pd
import numpy as np

mrkt_df = pd.read_csv('C:\\Users\\piked\\Documents\\Ffxiv Outputs\\Ffxiv Scripts\\Output Csv\\Hi_Mrkt_Vel v2.csv') \
    .groupby(['Item_Id', 'Name', 'GatheringType', 'World', 'Reg_Sale_Vel', 'NQ_Sale_Vel', 'HQ_Sale_Vel', 'Composite_Score', 'Total_Purchases', 'Avg_Unit_Price', 'RN', 'Buyer', 'Avg_Qty', 'Unit_Price', 'Qty', 'Overall_Cost', 'Outlier', 'HQ_Ind', 'Timestamp'], as_index=False) \
    .agg(GatheringLevel=pd.NamedAgg(column='GatheringLevel', aggfunc='min'))[['Item_Id', 'Name', 'GatheringType', 'GatheringLevel', 'World', 'Reg_Sale_Vel', 'NQ_Sale_Vel', 'HQ_Sale_Vel', 'Composite_Score', 'Total_Purchases', 'Avg_Unit_Price', 'RN', 'Buyer', 'Avg_Qty', 'Unit_Price', 'Qty', 'Overall_Cost', 'Outlier', 'HQ_Ind', 'Timestamp']]
# Convert to nullable float type
mrkt_df['GatheringLevel'] = mrkt_df['GatheringLevel'].apply(lambda x: int(x) if pd.notna(x) else x).astype('Int64')

print(f" Data Between: {mrkt_df['Timestamp'].min()}  - {mrkt_df['Timestamp'].max()}")
display(mrkt_df)
```

<h3>Gathering Analysis</h3>

```python
import pandas as pd
import numpy as np
import ipywidgets as widgets
from IPython.display import display, Markdown

# Varables to Change Pivots
startLevel = 1
endLevel = 100
GatheringType = 'all'

# Boolean to change pivot based on gathering type
if GatheringType == 'all':
    gathMrkt_df =  mrkt_df[(mrkt_df['Outlier'] != True) & (~mrkt_df['GatheringType'].isin(['Not Gathered','●銛'])) & (mrkt_df['GatheringLevel'] >= startLevel) & (mrkt_df['GatheringLevel'] <= endLevel)]
else:
    gathMrkt_df =  mrkt_df[(mrkt_df['Outlier'] != True) & (~mrkt_df['GatheringType'].isin(['Not Gathered','●銛'])) & (mrkt_df['GatheringLevel'] >= startLevel) & (mrkt_df['GatheringLevel'] <= endLevel) & (mrkt_df['GatheringType'] == GatheringType)]

# Totals Pivot
GathTtlPivot = gathMrkt_df.pivot_table(
    index=['GatheringType'],
    values=['Buyer','Overall_Cost', 'Reg_Sale_Vel'],
    aggfunc={'Buyer': 'count','Overall_Cost': 'sum', 'Reg_Sale_Vel': lambda x: round(x.mean(), 2)}) \
        .rename(columns={'Buyer': 'Total Transactions','Overall_Cost' : 'Total Profitability','Reg_Sale_Vel' : 'Avg. Reg Sale Vel'})
GathTtlPivot['Total Profitability'] = GathTtlPivot['Total Profitability'].apply(lambda x: f"${x:,.2f}")
# Reset the index for a flat table format
GathTtlPivot.reset_index(inplace=True)
GathTtlPivot = GathTtlPivot.rename(columns={'GatheringType' : 'Gathering Type'})

# By Gathering Level Pivot
GathLvlPivot = gathMrkt_df.pivot_table(
    index=['GatheringType', 'GatheringLevel'],
    values=['Buyer','Overall_Cost', 'Reg_Sale_Vel'],
    aggfunc={'Buyer': 'count','Overall_Cost': 'sum', 'Reg_Sale_Vel': lambda x: round(x.mean(), 2)}) \
        .rename(columns={'Buyer': 'Total Transactions','Overall_Cost' : 'Total Profitability','Reg_Sale_Vel' : 'Avg. Reg Sale Vel'})
GathLvlPivot['Total Profitability'] = GathLvlPivot['Total Profitability'].apply(lambda x: f"${x:,.2f}")
# Reset the index for a flat table format
GathLvlPivot.reset_index(inplace=True)
GathLvlPivot = GathLvlPivot.rename(columns={'GatheringType' : 'Gathering Type','GatheringLevel' : 'Level'})

GathItem = gathMrkt_df[['Item_Id', 'Name', 'GatheringType', 'GatheringLevel', 'Total_Purchases', 'Avg_Unit_Price', 'Avg_Qty', 'Overall_Cost', 'Reg_Sale_Vel']].copy()
GathItem['Total Profitability'] = GathItem.groupby('Item_Id')['Overall_Cost'].transform('sum').apply(lambda x: f"${x:,.2f}")
GathItem = GathItem[['Item_Id', 'Name', 'GatheringType', 'GatheringLevel', 'Total_Purchases', 'Avg_Unit_Price', 'Avg_Qty', 'Total Profitability', 'Reg_Sale_Vel']].drop_duplicates()

display(Markdown(f"# 2 Week Profitability by Gathering Type (Levels: {startLevel} - {endLevel})"))
display(GathTtlPivot)
display(Markdown(f"# 2 Week Profitability by Gathering Type & Level (Levels: {startLevel} - {endLevel})"))
display(GathLvlPivot)
display(Markdown(f"# 2 Week Profitability Granular (Levels: {startLevel} - {endLevel})"))
display(GathItem)
```
