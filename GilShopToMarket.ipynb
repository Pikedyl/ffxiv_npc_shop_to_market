Npc Resident API Pull

```python
import pandas as pd, requests, json, time, datetime, math

private_key = '29309ca885bb4c9096e08ef47213962d19137c17218b4f0a826024807837d435'
api_url = 'https://xivapi.com'
resident_data = '/ENpcResident'
pg_url = api_url + resident_data + "?private_key=" + private_key
#pg_ttl = requests.get(pg_url).json()["Pagination"]["PageTotal"]
pg_ttl = 2
ttl_npc = requests.get(pg_url).json()["Pagination"]["ResultsPerPage"]*pg_ttl

#Paginates to get all Npc Url
pg_processed = 0
url_list = []
print('Starting Page Request - ' + str(datetime.datetime.now().strftime('%Y-%m-%d %I:%M:%S %p')))
for page in range(1,pg_ttl+1):
    api = requests.get(api_url + resident_data + "?page=" + str(page)).json()["Results"]
    pg_processed += 1
    pg_prg = 'Page: ' + str(page) + ' - ' + str(math.ceil(100*(pg_processed/pg_ttl))) + '%'
    print(pg_prg, end='\r')
    if pg_processed%20==0: # Waits 2 sec every 20 requests
        time.sleep(2) 
    url_list = url_list + api
print('\nComplete - ' + str(datetime.datetime.now().strftime('%Y-%m-%d %I:%M:%S %p')))

#Formats Npc Urls
resi_url_df = pd.DataFrame(url_list) \
    #.iloc[:22]
resi_url_df['Url'] = api_url + resi_url_df['Url']
urls_list = resi_url_df['Url'].to_list()

#Check for and remove duplicate npc urls
uniq_urls_list = []
for url in urls_list:
    if url not in uniq_urls_list:
        uniq_urls_list.append(url)
if len(urls_list)-len(uniq_urls_list)>0:
    print("Duplicate Npc Url found and Removed: " + str(len(urls_list)-len(uniq_urls_list)))

#uniq_urls_list = 'https://xivapi.com/ENpcResident/1001787'.split(',')  #Bango Zango

#Create List of Vendors
row_processed = 0
shop_found = 0
npc_list = []
print('Starting Npc Request - ' + str(datetime.datetime.now().strftime('%Y-%m-%d %I:%M:%S %p')))
for url in uniq_urls_list:
    shop_npc_json = requests.get(url).json()
    shop_npc_df = pd.json_normalize(shop_npc_json)
    row_processed += 1
    npc_str = 'Npc: ' + str(row_processed) + ' - ' + str(math.ceil(100*(row_processed/len(uniq_urls_list)))) + '%'
    print(npc_str, end='\r')
    if len(shop_npc_df['Name'].values[0]) > 0:
        if len(shop_npc_df['GilShop'].values[0]) > 0:
            shop_npc_df = shop_npc_df[['ID','Name','Title','Base.Race.Name','GilShop']]
            shop_npc = shop_npc_df.values.tolist()
            npc_list = npc_list + shop_npc
            shop_found += 1
            print('Found: ' + str(shop_found) + ' Merchant Npc', end='\r')
    if row_processed%20==0: # Waits 2 sec every 20 requests
        time.sleep(2)
print('\nComplete - ' + str(datetime.datetime.now().strftime('%Y-%m-%d %I:%M:%S %p')))

output_df = pd.DataFrame(npc_list,columns=['Npc_Id','Npc_Name','Affiliate','Npc_Race','Gil_Shop']) \
    .explode('Gil_Shop').reset_index(drop=True)
output_df = output_df.merge(pd.json_normalize(output_df['Gil_Shop']), left_index=True, right_index=True).drop('Gil_Shop', axis=1)
output_df = pd.json_normalize(output_df[['Npc_Id','Npc_Name','Affiliate','Npc_Race','Name','Items']] \
    .rename(columns={'Name' : 'Item_Category'}).explode('Items').to_dict('records'))

GilShop_df = output_df[['Npc_Id','Npc_Name','Affiliate','Npc_Race','Item_Category','Items.ID','Items.Name_en','Items.StackSize','Items.PriceLow','Items.PriceMid','Items.Url']] \
    .rename(columns={'Items.ID' : 'Item_Id','Items.Name_en' : 'Item_Name','Items.StackSize' : 'Stack_Size', 'Items.PriceLow' : 'Price_Low', 'Items.PriceMid' : 'Price_Mid','Items.Url' : 'Item_Url'})
GilShop_df['Item_Url'] = api_url + GilShop_df['Item_Url']

#Remove redundant duplicate item id urls
item_urls_list = GilShop_df['Item_Url'].to_list()
uniq_item_urls_list = []
for url in item_urls_list:
    if url not in uniq_item_urls_list:
        uniq_item_urls_list.append(url)
if len(item_urls_list)-len(uniq_item_urls_list)>0:
    print("Duplicate Item Urls Found and Removed: " + str(len(item_urls_list)-len(uniq_item_urls_list)))

#Items api call
items_processed = 0
item_list = []
print('Starting Item Request - ' + str(datetime.datetime.now().strftime('%Y-%m-%d %I:%M:%S %p')))
for url in uniq_item_urls_list:
    item_json = requests.get(url).json()
    items_processed += 1
    item_tmp_df = pd.json_normalize(item_json)
    item_tmp_df = item_tmp_df[['ID','IsUntradable']]
    items = item_tmp_df.values.tolist()
    item_list = item_list + items
    item_str = 'Item: ' + str(items_processed) + ' - ' + str(math.ceil(100*(items_processed/len(uniq_item_urls_list)))) + '%'
    print(item_str, end='\r')
    if items_processed%20==0: # Waits 2 sec every 20 requests
        time.sleep(2)
print('\nComplete - ' + str(datetime.datetime.now().strftime('%Y-%m-%d %I:%M:%S %p')))
items_df = pd.DataFrame(item_list,columns=['Item_Id','Market_Prohibited'])
if len(items_df['Item_Id'].values)-len(items_df['Item_Id'][items_df['Market_Prohibited'] == 0].values)>0:
    print('Removed: ' + str(len(items_df['Item_Id'].values)-len(items_df['Item_Id'][items_df['Market_Prohibited'] == 0].values)) + ' Market Prohibited Items')

#Using inner join to remove Market Prohibited Items
GilShop_df = GilShop_df.merge(items_df[items_df['Market_Prohibited'] != 1],how='inner',on='Item_Id').drop(columns=['Item_Url'])



display(GilShop_df)
```
